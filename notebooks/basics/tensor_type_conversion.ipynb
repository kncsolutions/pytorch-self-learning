```python
import torch
```

```python
x = torch.tensor([1, 2, 3])
print("Original :", x.dtype)
print("Float :", x.float().dtype)
print("Double :", x.double().dtype)
print("Int :", x.int().dtype)
print("Long :", x.long().dtype)
print("Bool :", x.bool().dtype)
```

# Observe how type conversion is taking place

```python
x_to = x.to(torch.float64)
print("After using tensor.to() method the datatype is :", x_to.dtype)
```

* decimal computation

| float32  | lower  | moderate  | fast   |
| float64  | higher | very high | slower |

Deep learning usually uses:

float32

Scientific computing sometimes prefers:

float64

# tensor.to()

is the MOST general conversion method.

It can change:

* datatype
* device
* layout

```python
try:
    x_to = x.to("cuda")
    print("After using tensor.to() method to change device type the device is :", x_to.device)
except Exception:
    print("x"*50)
    print("Torch not compiled with CUDA enabled or GPU not available")
    print("x" * 50)
    pass
```

# Meaning
tensor.cuda()
Moves tensor into:
GPU memory
This enables:

* massively parallel computation
* neural network acceleration
* transformer training
* large tensor operations

---

# Important Insight

This single line:

x.cuda()

can change execution from:

* CPU kernels

to:

* CUDA GPU kernels

without changing your mathematical code.

That abstraction is one of PyTorch’s greatest innovations.

---

# Very Important Rule

Operations require tensors on SAME device.

This FAILS:

cpu_tensor + gpu_tensor

because:

* CPU memory
* GPU memory

are separate computational domains.
