```python
import torch
```

```python
BREAK_SIZE = 70
```

```python
####################################################################
# BASE TENSORS
####################################################################

x = torch.tensor([
    [1, 5, 3],
    [7, 2, 9]
])

y = torch.tensor([
    [1, 4, 8],
    [6, 2, 10]
])

print("Tensor x :")
print(x)

print("\nTensor y :")
print(y)
```

####################################################################
# torch.eq()
####################################################################


####################################################################
# DESCRIPTION:
#
# eq = equal
#
# Compares corresponding elements.
#
# Returns:
# True  -> if equal
# False -> otherwise
#
####################################################################

####################################################################
# MATHEMATICAL FORMULATION:
#
# z_ij = (x_ij == y_ij)
#
####################################################################

####################################################################
# BASIC ALGORITHM:
#
# for every element:
#
# if x_ij == y_ij:
#     True
# else:
#     False
#
####################################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.eq()")
z1 = torch.eq(x, y)

print(z1)
```

####################################################################
# torch.where()
####################################################################


####################################################################
# DESCRIPTION:
#
# Conditional selection operation.
#
# Selects values depending on condition.
#
####################################################################

####################################################################
# MATHEMATICAL FORMULATION:
#
# z_ij =
#
#     a_ij   if condition is True
#
#     b_ij   otherwise
#
####################################################################

####################################################################
# BASIC ALGORITHM:
#
# if condition:
#     select from tensor A
# else:
#     select from tensor B
#
####################################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.where()")
condition = x > y

print("Condition (x > y) :")
print(condition)

z7 = torch.where(condition, x, y)

print("\nSelected tensor :")
print(z7)
```

####################################################################
# PRACTICAL EXAMPLE
####################################################################



####################################################################
# ReLU-like behavior using torch.where()
####################################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("PRACTICAL EXAMPLE")
sample = torch.tensor([
    [-3.0, -1.0, 2.0],
    [4.0, -5.0, 6.0]
])

print("Original Tensor :")
print(sample)

relu_like = torch.where(
    sample > 0,
    sample,
    torch.tensor(0.0)
)

print("\nReLU-like Output :")
print(relu_like)
```
