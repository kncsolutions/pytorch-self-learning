```python
import torch
```

```python
BREAK_SIZE = 60
```

```python
print("=" * BREAK_SIZE)
print("PYTORCH MATHEMATICAL OPERATIONS PLAYGROUND")
print("=" * BREAK_SIZE)

print(f"PyTorch Version : {torch.__version__}")

print("=" * BREAK_SIZE)
```

############################################################
# BASE TENSORS
############################################################

```python
x = torch.tensor([
    [1.0, 2.0],
    [3.0, 4.0]
])

y = torch.tensor([
    [5.0, 6.0],
    [7.0, 8.0]
])

print("Tensor x :")
print(x)

print("\nTensor y :")
print(y)
```

############################################################
# torch.add()
# Element-wise addition.

# Mathematical Formula

# z_{ij} = x_{ij} + y_{ij}
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.add()")

z1 = torch.add(x, y)

print(z1)
```

############################################################
# What happened here?
# Was the addition element-wise?
############################################################

############################################################
# torch.sub()
# Element-wise subtraction.

# Mathematical Formula

# z_{ij} = x_{ij} - y_{ij}
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.sub()")

z2 = torch.sub(x, y)

print(z2)
```

############################################################
# Observe:
# Which tensor is subtracted from which?
############################################################

############################################################
# # torch.mul()
# Element-wise multiplication.

# Mathematical Formula

# z_{ij} = x_{ij} *. y_{ij}
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.mul()")

z3 = torch.mul(x, y)

print(z3)
```

############################################################
# Important:
# Is this matrix multiplication?
# Or element-wise multiplication?
############################################################

############################################################
# torch.div()
# Element-wise division.

# Mathematical Formula

# z_{ij} = {x_{ij}}./{y_{ij}}

############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.div()")

z4 = torch.div(y, x)

print(z4)
```

############################################################
# Observe:
# Division occurred element by element.
############################################################

############################################################
# torch.matmul()
# Matrix multiplication.

# Mathematical Formula

# C{i,j} = summation(k) (A{i, k} * B{k, j})

############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.matmul()")

z5 = torch.matmul(x, y)

print(z5)
```

############################################################
# Important:
# This is REAL matrix multiplication.
############################################################

############################################################
# torch.mm()
# 2D matrix multiplication only.

# Formula

# Same as matrix multiplication:

# C{i,j} = summation(k) (A{i, k} * B{k, j})
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.mm()")

z6 = torch.mm(x, y)

print(z6)
```

############################################################
# Question:
# Compare torch.matmul() vs torch.mm()
############################################################

############################################################
# torch.bmm()
Batch matrix multiplication.

Mathematical Formula

C^{(b)} = A^{(b)}B^{(b)}
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.bmm()")

a = torch.rand(2, 2, 3)

b = torch.rand(2, 3, 2)

print("Tensor a shape :", a.shape)
print("Tensor b shape :", b.shape)

z7 = torch.bmm(a, b)

print("Result shape :", z7.shape)

print(z7)
```

############################################################
# Important:
# bmm() = batch matrix multiplication
############################################################

############################################################
# torch.exp()
y = e^x
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.exp()")

z8 = torch.exp(x)

print(z8)
```

############################################################
# Observe:
# e^x computed element-wise.
############################################################

############################################################
# torch.log()
y = ln(x)
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.log()")

z9 = torch.log(x)

print(z9)
```

############################################################
# Observe:
# Natural logarithm applied element-wise.
############################################################

############################################################
# torch.sqrt()
y = sqrt(x)
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.sqrt()")

z10 = torch.sqrt(x)

print(z10)
```

############################################################
# torch.pow()
y = x .^ 2
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.pow()")

z11 = torch.pow(x, 2)

print(z11)
```

############################################################
# Observe:
# Each element squared.
############################################################

############################################################
# torch.abs()
y = |x|
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.abs()")

neg_tensor = torch.tensor([
    [-1.0, -2.0],
    [3.0, -4.0]
])

print("Original tensor :")
print(neg_tensor)

z12 = torch.abs(neg_tensor)

print("Absolute tensor :")
print(z12)
```

############################################################
# torch.sin()
y = sin(x)
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.sin()")

z13 = torch.sin(x)

print(z13)
```

############################################################
# torch.cos()
y = cos(x)
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.cos()")

z14 = torch.cos(x)

print(z14)
```

############################################################
# torch.tanh()
Hyperbolic tangent activation.

Mathematical Formula

tanh(x)={e^x-e^{-x}}/{e^x+e^{-x}}
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.tanh()")

z15 = torch.tanh(x)

print(z15)
```

############################################################
# Important:
# tanh() is a very important activation function.
Important Insight

tanh() compresses values between:

-1 and 1

Historically important in:

recurrent neural networks
signal processing

############################################################

############################################################
# torch.relu()
Rectified Linear Unit.

Mathematical Formula

ReLU(x)=max(0,x)
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.relu()")

relu_input = torch.tensor([
    [-2.0, -1.0],
    [3.0, 4.0]
])

print("Input tensor :")
print(relu_input)

z16 = torch.relu(relu_input)

print("ReLU output :")
print(z16)
```

############################################################
# Observe:
# Negative values become zero.
Important Insight

ReLU is one of the MOST important functions in deep learning.

Why?

simple
computationally efficient
avoids vanishing gradients better than sigmoid
############################################################

############################################################
# torch.sigmoid()
Mathematical Formula

sigma(x)={1}/ {1+e^{-x}}
############################################################

```python
print("\n" + "#" * BREAK_SIZE)
print("method torch.sigmoid()")

z17 = torch.sigmoid(x)

print(z17)
```

############################################################
# Important:
# Sigmoid compresses values between 0 and 1.


Useful in:

probabilities
binary classification
logistic regression
