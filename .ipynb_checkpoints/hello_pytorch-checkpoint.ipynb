# Hello PyTorch

This notebook demonstrates the "Hello World" equivalent in PyTorch.

Instead of printing simple text only, PyTorch introduces the concept of **tensors**, which are multidimensional computational objects used for numerical computation, machine learning, and deep learning.

---

# Importing PyTorch

```python
import torch
```

The `torch` module is the core library of PyTorch.

It provides:

* tensor operations
* automatic differentiation
* GPU acceleration
* neural network utilities
* backend numerical computation engines


````python


def hello_pytorch():
    print("="*30)
    print("Hello PyTorch")
    print("="*30)
    print("Pytorch Version : ", torch.__version__)
    print("Simple basic operations with pytorch: ")
    x = torch.tensor([1, 2, 3])
    print("Your first tensor : ", x)
    
    y = torch.tensor([4, 5, 6])
    print("Your second tensor : ", y)
    z = x + y
    print("Your third tensor after summing x and y : ", z)

    print("=" * 30)
    print("Let's see some features of x :")
    print("=" * 30)

    print("Type of x : ", type(x))
    print("Data type of x : ", x.dtype)
    print("Shape of x : ", x.shape)
    print("Device of x : ", x.device)
````

# Hello PyTorch

This notebook demonstrates the "Hello World" equivalent in PyTorch.

Instead of printing simple text only, PyTorch introduces the concept of **tensors**, which are multidimensional computational objects used for numerical computation, machine learning, and deep learning.

---

# Importing PyTorch

```python
import torch
```

The `torch` module is the core library of PyTorch.

It provides:

* tensor operations
* automatic differentiation
* GPU acceleration
* neural network utilities
* backend numerical computation engines

---

# First Tensor

```python
x = torch.tensor([1, 2, 3])

print(x)
```

## Output

```python
tensor([1, 2, 3])
```

This creates a **1-dimensional tensor** (vector tensor).

Internally, PyTorch:

1. allocates memory
2. stores numerical values
3. infers datatype
4. creates tensor metadata
5. prepares backend execution structures

---

# Understanding Tensor Properties

```python
print(type(x))
print(x.dtype)
print(x.shape)
print(x.device)
print(x.ndim)
```

## Expected Output

```python
<class 'torch.Tensor'>

torch.int64

torch.Size([3])

cpu

1
```

---

# Meaning of Each Property

| Property  | Meaning              |
| --------- | -------------------- |
| `type(x)` | Object type          |
| `dtype`   | Numerical datatype   |
| `shape`   | Tensor dimensions    |
| `device`  | CPU or GPU location  |
| `ndim`    | Number of dimensions |

---

# Tensor Arithmetic

```python
y = torch.tensor([4, 5, 6])

z = x + y

print(z)
```

## Output

```python
tensor([5, 7, 9])
```

PyTorch performs optimized vectorized computation internally.

You do not manually implement:

* loops
* memory allocation
* vector addition algorithms

PyTorch automatically dispatches computation to optimized backend kernels.

---

# Why Tensors Matter

Everything in modern AI eventually becomes tensors:

* images
* embeddings
* audio
* latent vectors
* transformer attention matrices
* semantic representations
* market state representations

PyTorch tensors are computational structures inspired by mathematical tensors and optimized for differentiable numerical computation.

---

# Mental Model

```text
Python Code
    ↓
PyTorch Tensor API
    ↓
Tensor Operations
    ↓
Backend Dispatcher
    ↓
CPU / GPU Kernels
    ↓
Numerical Computation
```

This abstraction is one of the core strengths of PyTorch.
