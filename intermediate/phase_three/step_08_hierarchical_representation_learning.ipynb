    intermediate interaction patterns

    higher-order representations

    semantic structure

    hierarchical representation learning

```python

import torch
import torch.nn as nn
import torch.nn.functional as F

torch.manual_seed(42)
```

# 1. Creating a Simple Spatial Tensor

We create a simple geometric structure.

The tensor contains:

- local regions
- strong contrast
- structured patterns

This helps visualize:

    representation evolution

```python

x = torch.tensor([
    [0., 0., 1., 1., 1., 0., 0.],
    [0., 1., 1., 1., 1., 1., 0.],
    [1., 1., 5., 5., 5., 1., 1.],
    [1., 1., 5., 10., 5., 1., 1.],
    [1., 1., 5., 5., 5., 1., 1.],
    [0., 1., 1., 1., 1., 1., 0.],
    [0., 0., 1., 1., 1., 0., 0.]
])

print("Input Tensor:")
print(x)

print("\nInput Shape:")
print(x.shape)
```

# 2. Converting to Conv2D Format

PyTorch Conv2D expects:

    [batch, channels, height, width]

```python

x_conv = x.reshape(1, 1, 7, 7)

print("Conv Tensor Shape:")
print(x_conv.shape)
```

# 3. First Representation Layer

The first layer learns:

    local interaction structure

This layer typically captures:

- local contrast
- edges
- primitive patterns

We use a simple interaction kernel.

```python

conv1 = nn.Conv2d(
    in_channels=1,
    out_channels=2,
    kernel_size=3,
    padding=1
)
```

# 4. Running First Convolution Layer

The tensor now transforms into:

    shallow feature representations

Each output channel becomes:

    a different response field

```python

y1 = conv1(x_conv)

print("Layer 1 Output Shape:")
print(y1.shape)
```

# Understanding Layer 1

Input shape:
    [1, 1, 7, 7]

Output shape:
    [1, 2, 7, 7]

Important observation:

The network already generated:

    multiple representation perspectives

Each channel learns different:

- interaction sensitivities
- local response patterns

# 5. Applying Non-Linearity

Without non-linearity:

    stacked layers collapse into

Activation functions introduce:

    representation flexibility

This allows networks to learn:

    complex interaction geometry
Activation Function?
# Activation Function Intuition

If you are new to AI, many terms may initially
sound mysterious or magical.

Actually they are not.

An activation function determines how strongly
a neuron response should remain active after
a computation is performed.

---

# What Is an Artificial Neuron?

A neuron performs a weighted combination:

z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b

Where:

 x = inputs
 w = learned weights
 b = bias term

This operation is fundamentally:


a learned linear interaction


Convolution layers perform a structured spatial
version of this same idea using shared kernels.

So conceptually you already understand a major part
of neural computation.

---

# What Does "Firing" Mean?

After the weighted computation is performed,
the network decides how much of the signal
should remain active.

This is where the activation function operates.

For example:

if signal is strong enough:
preserve it

otherwise:
suppress it

This is only an intuition,
not the exact implementation.

---

# ReLU Example

In this module we use:

\mathrm{ReLU}(x)=\max(0,x)

Meaning:

* positive values remain active
* negative values become zero

So ReLU behaves like a simple gating mechanism.

It allows some signals to propagate while suppressing others.

---

# Why Non-Linearity Matters

Without activation functions:

multiple stacked layers collapse into
one large linear transformation.

Non-linearity allows neural networks to learn:

* complex patterns
* curved decision boundaries
* hierarchical representations
* rich interaction geometry

This is one of the foundational reasons
deep learning works.



```python

a1 = F.relu(y1)

print("Activated Layer 1 Shape:")
print(a1.shape)
```

# 6. Second Representation Layer

Now the second layer no longer sees:

    raw spatial values

Instead it sees:

    Layer 1 representations

Thus abstraction begins increasing.

```python

conv2 = nn.Conv2d(
    in_channels=2,
    out_channels=4,
    kernel_size=3,
    padding=1
)
```

# 7. Running Second Convolution Layer

The second layer aggregates:

    previously learned representations

This creates:

    more abstract interaction structures

```python

y2 = conv2(a1)

a2 = F.relu(y2)

print("Layer 2 Output Shape:")
print(a2.shape)
```

# Understanding Layer 2

Layer 2 representations become:

- richer
- more distributed
- more abstract

The system now combines:

    multiple shallow interaction patterns

to form more complex structures.

# 8. Third Representation Layer

Now abstraction deepens further.

The third layer observes:

    higher-order representations

rather than simple local structures.

```python

conv3 = nn.Conv2d(
    in_channels=4,
    out_channels=8,
    kernel_size=3,
    padding=1
)
```

# 9. Running Third Convolution Layer

Representation hierarchy deepens.

The network gradually builds:

    broader contextual understanding

```python

y3 = conv3(a2)

a3 = F.relu(y3)

print("Layer 3 Output Shape:")
print(a3.shape)
```

# Important Observation

Representation depth increased:

Layer 1:
    2 feature maps

Layer 2:
    4 feature maps

Layer 3:
    8 feature maps

The network progressively expands:

- interaction diversity
- representation capacity
- abstraction complexity

```python

print("Layer 1 Shape:")
print(a1.shape)

print("\nLayer 2 Shape:")
print(a2.shape)

print("\nLayer 3 Shape:")
print(a3.shape)
```
